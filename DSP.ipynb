##  Brute force solution

```python
%%time
function_min=1000000
for x in np.arange(-10, 10, 0.01):
    for y in np.arange(-10, 10, 0.01):
        function = np.exp(x+y)+np.exp(-x-y)+np.exp(-x+y+4)+np.exp(x-y-4)
        if (function<function_min):
            function_min=function
            x_minum=x
            y_minum=y
print ("minmum cost = " + str(function_min))
print ("x = " + str(x_minum))
print ("y = " + str(y_minum))
```

## gradient descent

```python
X=0
Y=0
```

```python
# GRADED FUNCTION: optimize

def optimize_gradient( x, y, learning_rate , x_minum , y_minum):
    
    Function = []
    i=1
    while np.sqrt((x-x_minum)**2+(y-y_minum)**2) > 0.001:
        Function = np.exp(x+y)+np.exp(-x-y)+np.exp(-x+y+4)+np.exp(x-y-4)
        dx =np.exp(x+y)-np.exp(-x-y)-np.exp(-x+y+4)+np.exp(x-y-4)
        dy =np.exp(x+y)-np.exp(-x-y)+np.exp(-x+y+4)-np.exp(x-y-4)
        Function = np.array(Function)
        x = x-learning_rate*dx
        y = y-learning_rate*dy
        print ("iteration %i: %f" %(i, Function))
        i=i+1

    
    return x ,y
```

```python
%%time
x_opt,y_opt= optimize_gradient( X, Y, learning_rate = 0.002 ,x_minum= x_minum ,y_minum=y_minum)
print ("x = " + str(x_opt))
print ("y = " + str(y_opt))
```

```python
%%time
x_opt,y_opt= optimize_gradient( X, Y, learning_rate = 0.02 ,x_minum= x_minum ,y_minum=y_minum)

print ("x = " + str(x_opt))
print ("y = " + str(y_opt))
```

## Newton method

```python
def huissen(x ,y):
        dx =np.exp(x+y)-np.exp(-x-y)-np.exp(-x+y+4)+np.exp(x-y-4)
        dy =np.exp(x+y)-np.exp(-x-y)+np.exp(-x+y+4)-np.exp(x-y-4)
        dxy =np.exp(x+y)+np.exp(-x-y)-np.exp(-x+y+4)-np.exp(x-y-4)
        dxx =np.exp(x+y)+np.exp(-x-y)+np.exp(-x+y+4)+np.exp(x-y-4)
        dyx =np.exp(x+y)+np.exp(-x-y)-np.exp(-x+y+4)-np.exp(x-y-4)
        dyy =np.exp(x+y)+np.exp(-x-y)+np.exp(-x+y+4)+np.exp(x-y-4)
        return dx,dy,dxy,dxx,dyx,dyy
```

#### A square matrix is called positive definite if it is symmetric and all its eigenvalues λ are positive, that is λ > 0. 
#### Because these matrices are symmetric, the principal axes theorem plays a central role in the theory.
#### If A is positive definite, then it is invertible and det A > 0.

```python
# GRADED FUNCTION: optimize

def optimize_Newoten(x, y, x_minum , y_minum, learning_rate=1):
    
    Function = []
    i=1
    while np.sqrt((x-x_min)**2+(y-y_min)**2) > 0.001:
        Function = np.exp(x+y)+np.exp(-x-y)+np.exp(-x+y+4)+np.exp(x-y-4)
        dx,dy,dxy,dxx,dyx,dyy=huissen(x ,y)
        N=np.linalg.inv(np.array([[dxx,dxy],[dyx,dyy]])) @ np.array([[dx],[dy]])
        Function = np.array(Function)
        x = x-learning_rate*N[0]
        x=x[0]
        y = y-learning_rate*N[1]
        y=y[0]
        print ("iteration %i: %f" %(i, Function))
        i=i+1
    
    return x ,y
```

```python
%%time
x_opt,y_opt= optimize_Newoten( x=1, y=1, x_minum= x_minum ,y_minum=y_minum)

print ("x = " + str(x_opt))
print ("y = " + str(y_opt))
```
